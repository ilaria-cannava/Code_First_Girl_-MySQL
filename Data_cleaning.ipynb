```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df=pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
# Get the data types of each column
column_types = df.dtypes

# Select only the columns with categorical data types
categorical_columns = column_types[column_types == 'object'].index.tolist()

# Print the list of categorical columns
print("Categorical Columns:")
for column in categorical_columns:
    print(column)
```

```python
# Get the list of categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Print unique values for each categorical column
for column in categorical_columns:
    unique_values = df[column].unique()
    print(f"Unique values for {column}:")
    for value in unique_values:
        print(value)
    print()  # Add a blank line after printing unique values for each column
```

```python
unique_values = df['Education'].unique()
print(unique_values)
```

```python
# Check for null values in the DataFrame
null_values = df.isnull()

# Count the number of null values in each column
null_counts = null_values.sum()

print(null_counts)
```

```python
# Creating a palette
palette = sns.color_palette(['#10b5b6','#6b3655', '#ECA400','#026c7c', '#f76464', '#f20d0d', '#230903'])
sns.palplot(palette)
sns.set_palette(['#ECA400','#026c7c', '#f76464', '#f20d0d', '#230903'])
```

```python
# Visualising the target
plt.style.use('default')
plt.figure(figsize=(5,5))   
ax = sns.countplot(data=df, x=df['ATTRITION'], palette=palette)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.set_xticklabels(['Employees','Leavers'])
plt.show()
```

```python
sns.countplot(x='ATTRITION', data=df, hue='Gender')
plt.legend(labels=["Female","Male"])
plt.xlabel('')
plt.ylabel('')
plt.xticks(ticks=np.arange(0,2,1), labels=["Employees", "Leavers"])
plt.show()
```
